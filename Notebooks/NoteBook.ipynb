```python
# Install required packages
#!pip install ultralytics opencv-python-headless

import cv2
from ultralytics import YOLO
#from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
import time

# Load the best model
model = YOLO('runs/detect/train6/weights/best.pt')

# Open the video file
video_path = 'Cars- Busy Streets- City Traffic - No Copyright Royalty Free Stock Videos.mp4'  # Replace with your video path
cap = cv2.VideoCapture(video_path)

# Get video properties
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))

# Define the codec and create VideoWriter object
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter('/content/gdrive/MyDrive/dataset-1/output_video.mp4', fourcc, fps, (width, height))

# Process the video
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    
    # Run YOLOv8 inference on the frame
    results = model(frame)
    
    # Visualize the results on the frame
    annotated_frame = results[0].plot()
    
    # Write the frame into the output file
    out.write(annotated_frame)
    
    # Display the frame
    plt.imshow(cv2.cvtColor(annotated_frame, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()
    
    # Clear the output to show only the current frame
    from IPython.display import clear_output
    clear_output(wait=True)
    
    # Control the frame rate for live display
    time.sleep(1/fps)

# Release everything when job is finished
cap.release()
out.release()

print("Video processing complete. Output saved as 'output_video.mp4'")


```
